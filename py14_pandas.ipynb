## indexing, slicing

* `iloc` : integer location정수 인덱스(ndarray의 기본 인덱스) 기반으로 원소 참조 또는 슬라이싱
*`loc` : location. 레이블 기반의 원소 참조 또는 슬라이싱

```python
np.random.seed(1)
s1 = pd.Series(data=np.random.randint(-10,10, size=6))
s1
```

```python
#iloc를 사용한 인덱싱
print(s1.iloc[0]) # 시리즈에서 첫번째 원소
print(s1.iloc[-1]) # 시리즈에서 마지막 원소
```

```python
#iloc를 사용한 슬라이싱
print(s1.iloc[:3]) # 시리즈에서 첫 3개 원소 자르기
print(s1.iloc[-3:]) # 시리즈에서 마지막3개 원소 자르기
```

```python
# loc를 사용한 인덱싱
print(s1.loc[0])#series의 첫번째 원소
print(s1.loc[5]) #시리즈의 마지막 원소 음수 인덱스 사용 불가능!
```

```python
#loc를 사용한 슬라이싱
print(s1.loc[:2])
```

* boolean indexcing을 사용할 떄는, 파이썬의 논리 연산자(`and, or, not`)를 사용할 수 없음.!
* boolean indexcing에서는 `&, | ,~` 연산자를 사용함!
* 조건식과 조건식은 ()를 사용해서 구분 ,연산의 순서를 명시해야힘.

```python
# city가 '서울' 이거나, pop이 0.65이상인 행들을 선택:
df[(df['city']== '서울') | (df['pop'] <= 0.65)] # 첫번쨰 조건이거나 두번째 조건이거나.
```

```python
#경기도의 2023년 자료를 선택
df[(df.city == '경기')  &  (df.year == 2023)]
```

```python
#수학, 영어, 과학 컬럼의 기술 통계량의 출력df[(df['city'] == '서울') | (df['pop'] <= 0.65)]
exam[['math','english','science']].describe()
```

```python
# 1반 학생들의 데이터를 출력
exam[(exam['class'] == 1)]
```

```python
#* 수학의 평균을 찾고 출력

#exam[['math']].mean()
exam['math'].mean()
```

```python
#* 수학 점수가 평균 이상인 학생들의 데이터를 출력
exam[exam['math'] >= exam['math'].mean() ]
```

```python
#* 1반 학생들의 수학 점수 평균을 찾고 출력
exam[exam['class'] == 1].math.mean()
#* 2학 학생들의 수학 점수 평균을 찾고 출력
exam[exam['class'] == 2].math.mean()
```

```python
#* 반별 (class 별) 수학 점수 평균을 찾고 출력
for x in range(1,6):
    math_mean = exam[exam['class'] == x].math.mean()
    print(f'class {x} : {math_mean}')
```

```python
#* 반별 (class 별) 수학 점수 평균을 찾고 출력 새로운 방법
exam_dict = {'class': np.arange(1,6)}
means = []
for x in range(1,6):
    math_mean = exam[exam['class'] == x].math.mean()
    means.append(math_mean)
print(means)

exam_dict = {
    'class': np.arange(1,6),
    'math_mean': means
}
print(exam_dict)

df = pd.DataFrame(exam_dict)
df
```

```python
#* 세 과목의 점수가 모두 평균 이상인 학생들의 데이터를 출력
exam[(exam.math >= exam.math.mean()) & (exam.english >= exam.english.mean()) & (exam.science >= exam.science.mean())]
```

```python
#1반 학생들의 수학 평균점수 이상인 학생들
#exam[(exam.math >=(exam[exam['class'] == 1].math.mean()))]

math_mean_class1 = exam[exam['class'] == 1].math.mean()
exam[exam.math >= math_mean_class1]
```
